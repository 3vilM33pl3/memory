# Memory Layer — Python quickstart

Needs a running service (`docker compose up` or `memory service run`) and the demo corpus (`memory demo`).

```python
import sys; sys.path.insert(0, '../src')  # or: pip install memory-layer-client
from memory_layer import MemoryLayerClient

client = MemoryLayerClient()  # http://127.0.0.1:4040
client.health()
```

## Ask a question — cited answer or honest refusal

```python
answer = client.query('demo', 'How does reinforcement work?', deterministic=True)
print(answer.answer)
print('confidence:', answer.confidence, '| refused:', answer.insufficient_evidence)
[(r.summary, round(r.score, 1)) for r in answer.results[:3]]
```

## Record a durable fact (capture + bounded curation)

```python
client.remember('demo', title='Notebook run',
                summary='The Python client quickstart ran end to end.',
                tags=['python-client'])
```

## The memory graph, with decayed ACT-R activation per node

```python
graph = client.memory_graph('demo')
hot = sorted((n for n in graph['nodes'] if n.get('activation')),
             key=lambda n: -n['activation'])[:5]
[(round(n['activation'], 2), n['label'][:60]) for n in hot]
```
